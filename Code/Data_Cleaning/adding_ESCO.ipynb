```python
def getISCOjobCodes(job_title):
    url = f"https://ec.europa.eu/esco/api/search?language=pt&type=occupation&text={job_title}"
    payload={}
    headers = {}
    
    try:
        response = requests.request("GET", url, headers=headers, data=payload).json()
    except Exception as e:
        print(e)
        print(requests.request("GET", url, headers=headers, data=payload))

    try:
        title = response['_embedded']['results'][0]['title']
    except:
        title = ''

    if title == job_title:
        match = 'Yes'
    else:
        match = 'No'
    try:
        output = {'searchHit': title,
        'uri': response['_embedded']['results'][0]['uri'],
        'code': response['_embedded']['results'][0]['code'],
        'Exact Match': match
        }
    except:
        output = {'searchHit': title,
        'uri': '',
        'code': '',
        'Exact Match': match
        }

        # get ESCO skills
        # url = f"https://ec.europa.eu/esco/api/resource/occupation?uri={output['uri']}&language=pt"

        # response = requests.request("GET", url, headers=headers, data=payload).json()

        # output['broaderIscoGroup_title'] = response["_links"]['broaderIscoGroup']['title']
        # output['broaderIscoGroup_code'] = response["_links"]['broaderIscoGroup']['code']
        # output['Isco_skills'] = [i for i in response['_links']['hasEssentialSkill']['title']]

    return output

new_list = []
for job_title in tqdm(df['job_title']):
    time.sleep(.5)
    new_list.append(getISCOjobCodes(job_title))

new_df = pd.DataFrame(new_list)
new_df.head()
```
